# Dimensionality Reduction Low

```python
import dimensionality_reduction_utils as dru
from sklearn.decomposition import NMF
```

```python
n_components = 3

pca_data = dru.pca_analysis(X,
                            n_components=n_components,
                 plot_sqrt_eigvals=False,
                 plot_perc_variance_explained=True)
X_pca_low = pca_data["data_proj"]

model = NMF(n_components=n_components, init='random', random_state=0)
X_nmf_low = model.fit_transform(X)
```

```python
n_components = 10

pca_data = dru.pca_analysis(X,
                            n_components=n_components,
                 plot_sqrt_eigvals=False,
                 plot_perc_variance_explained=True)
X_pca_med = pca_data["data_proj"]

model = NMF(n_components=n_components, init='random', random_state=0)
X_nmf_med = model.fit_transform(X)
```

```python
def accuracy_cluster(model,y=labels_idx):
    return np.sum(model.labels_ != y)/len(y)
```

# Measuring the Accuracy of Different Clustering Methods

```python
"""
Purpose: To implement different methods of clustering
and measuing the accuracy (in combination with different preprocessing steps)

Pseudocode: 
For each model
    For each data preprocess step
        a) Train the model
"""
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.cluster import AgglomerativeClustering, FeatureAgglomeration
import pandas as pd
from sklearn.cluster import SpectralClustering


n_clusters = 4

data_preprocessed = dict(
full_dimension = X,
    pca_3_comp = X_pca_low,
    pca_10_comp = X_pca_med,
    nmf_3_comp = X_nmf_low,
    nmf_10_comp = X_nmf_med,
)

model_types = dict(
kmeans = KMeans(n_clusters=n_clusters,
                random_state=0),
    
hc_ward = AgglomerativeClustering(
    n_clusters=n_clusters,
    linkage="ward"),

hc_average = AgglomerativeClustering(
    n_clusters=n_clusters,
    linkage="average"),

hc_single = AgglomerativeClustering(
    n_clusters=n_clusters,
    linkage="single"),

hc_complete = AgglomerativeClustering(
    n_clusters=n_clusters,
    linkage="complete"),
    
spectral_clust_10_neighbor = SpectralClustering(n_clusters=n_clusters,
        assign_labels='discretize',
        n_neighbors = 10,
        random_state=0),
    
spectral_clust_5_neighbor = SpectralClustering(n_clusters=n_clusters,
        assign_labels='discretize',
        n_neighbors=5,
        random_state=0)
    
)
```

```python
verbose = True

model_accuracy = []

for clf_name,clf in model_types.items():
    if verbose:
        print(f"Model: {clf_name}")
    local_accuracy = dict(model=clf_name)
    for pre_name,pre_data in data_preprocessed.items():
        st = time.time()
        clf.fit(pre_data)
        local_accuracy[pre_name] = accuracy_cluster(clf)
        if verbose:
            print(f"Data name: {pre_name}: {time.time() - st}")
    model_accuracy.append(local_accuracy)    
```

```python
pd.DataFrame.from_records(model_accuracy)
```
