```r
library(MASS)
library(ISLR)
library(car)
library(leaps)
library(glmnet)
```

# Ridge and Lasso

## Ridge

### Hitters dataset
$y =$ salary

```r
names(Hitters)
```

```r
x=model.matrix(Salary~.,Hitters)[,-1]
y=na.omit(Hitters$Salary)

#Split into test and training
set.seed(1)
train=sample(1:nrow(x), nrow(x)/2)
test=(-train)
y.test=y[test]

# Ridge Regression
grid=exp(1)^seq(10,-1,length=100)
ridge.mod=glmnet(x[train,],y[train],alpha=0,lambda=grid)

plot(ridge.mod, "lambda")
```

## Lasso

```r
# The Lasso
grid=exp(1)^seq(10,-1,length=100)
lasso.mod=glmnet(x[train,],y[train],alpha=1,lambda=grid)
plot(lasso.mod, "lambda")
```

**Differences:**
- Ridge has non-zero coefficients for each $\lambda$
- Lasso has many zero coefficients

# Elastic Net

The loss function for elastic net is:

$$ L(\beta; \alpha, \lambda) = \| y - X\beta \|_2^2 + \lambda P(\beta; \alpha)$$ with
$$ (0\le\alpha\le 1)$$
for 
$$ P(\beta; \alpha) = (1-\alpha)/2||\beta||_2^2+\alpha||\beta||_1.$$

For more information, read [Regularization Paths for Generalized Linear Models
via Coordinate Descent](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2929880/)

```r
# The Lasso
grid=exp(1)^seq(10,-1,length=100)
lasso.mod = glmnet(x[train,],y[train],alpha=0.0,lambda=grid)
plot(lasso.mod, "lambda")
```

# Cross validation with Lasso

```r
alpha = 1.0

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=3)
plot(cvfit, main="MSE for nfolds = 3")

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=10)
plot(cvfit, main="MSE for nfolds = 10")

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=length(train))
plot(cvfit, main="MSE for nfolds = n = 131 (LOOCV)")
```

# Cross validation with Ridge

```r
alpha = 0.0

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=5)
plot(cvfit, main="MSE for nfolds = 3")

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=10)
plot(cvfit, main="MSE for nfolds = 10")

cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=length(train))
plot(cvfit, main="MSE for nfolds = n = 131 (LOOCV)")
```

# Cross validation with elastic net

```r
alpha = 0.01
grid=exp(1)^seq(10,-1,length=100)
cvfit = cv.glmnet(x[train,], y[train], alpha = alpha,lambda=grid, nfolds=10)
plot(cvfit, main="MSE for nfolds = 10")
```

# Relaxed elastic net

```r
alpha = 0.1
cvfit = cv.glmnet(x[train,], y[train], alpha = alpha, lambda=grid, nfolds=10, relax=TRUE)
plot(cvfit, main="MSE for nfolds = 10")
```

If relax=TRUE then the values of gamma are used to mix the fits. If $\beta$ is the fit for lasso/elastic net, and $\beta_R$ is the relaxed fit (with unpenalized coefficients), then a relaxed fit mixed by $\gamma$ is $$\beta(\gamma)=(1-\gamma)\beta_R+\gamma\beta.$$ There is practically no extra cost for having a lot of values for gamma. However, 5 seems sufficient for most purposes. CV then selects both gamma and lambda.
