# Dimensionality Reduction

```python
import pandas_ml as pdml
import dimensionality_reduction_ml as dru
import pandas_utils as pu
```

```python
X,y = pdml.X_y(df,target_name = "id")
X
```

# --- Running on all excitatory and inhibitory cells --

```python
import numpy as np
labels_str = X["cell_type"].to_numpy()
(labels_str_unique,
 idx_to_name_map,
 labels_idx) = np.unique(labels_str,
                         return_index=True,
                         return_inverse=True)
str_order = labels_str[idx_to_name_map]
str_to_idx_map = text_to_plot_authors = {s:l_i for l_i,s in 
                        enumerate(str_order)}

# labels_colors = np.array(
#     mlu.generate_non_randon_named_color_list(
#         len(labels_str_unique))
# )
print(f"labels_str_unique = {labels_str_unique}")
# print(f"labels_colors= {labels_colors}")

str_to_idx_map
```

```python
X_data = pu.delete_columns(X,["cell_type"])
X_data
```

```python
import preprocessing_ml as preml
X_data= preml.scale_df(X_data)
X_data
```

# PCA

```python
n_components = 20
```

```python
pca_data = dru.pca_analysis(X_data,
                            n_components=n_components,
                 plot_sqrt_eigvals=False,
                 plot_perc_variance_explained=True,
                           whiten = True)
pca_data.keys()
```

```python
dru.plot_projected_data(
    pca_data["data_proj"][:,:3],
    labels_idx,
    cmap = "viridis",
text_to_plot_dict=str_to_idx_map
)
```

```python
dru.plot_projected_data_2D(pca_data["data_proj"][:,:2],
    labels_idx,
    cmap = "viridis",
text_to_plot_dict=str_to_idx_map)
```

# NMF

```python
from sklearn.decomposition import NMF
model = NMF(n_components=2, init='random', random_state=0)

W = model.fit_transform(preml.non_negative_df(X_data))
H = model.components_

dru.plot_projected_data_2D(
    W,
    labels=labels_idx,
text_to_plot_dict=str_to_idx_map)
```

# TSNE

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=3)
#returns the transformed data
W = model.fit_transform(X_data)

dru.plot_projected_data(
    W,
    labels=labels_idx,
    text_to_plot_dict=str_to_idx_map
)
```
