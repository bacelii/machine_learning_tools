```python
name_map = {
"Type I":"Type 1",
"Type 1 ": "Type 1",
'Breast Cancer ': 'Breast Cancer',
'Diffuse Large B Cell Lymphoma':'Diffuse Large B-Cell Lymphoma',
'Non Small Cell Lung Cancer ':'Non Small Cell Lung Cancer',
'Small Cell Carcinoma of  Lung':'Small Cell Carcinoma of Lung',
'Small Cell Carcinoma of the Lung':'Small Cell Carcinoma of Lung',
'Small Cell Carcinoma of Lung ':'Small Cell Carcinoma of Lung',
'Small Cell Lung Cancer':'Small Cell Carcinoma of Lung',
'Squamous Cell Carcinoma of Lung ':'Squamous Cell Carcinoma of Lung',
'Squmous Cell Carcinoma of Lung':'Squamous Cell Carcinoma of Lung',
'Thymic carcinoma':'Thymic Carcinoma',
}
name_map
```

```python
df = pd.read_excel("./SVC Stent_Cleaned up Data.xlsx")

import numpy as np
search_terms = ["Ethn", 
 "Age at Pro", 
 "Clinical Grade", 
 "Emergent Indication", 
 "ECOG",
"Etiology of SVC",
"Stenosis",
"Stanford",
"Azizi Classification",
"Azizi Type",
"Azizi Grade", 
"Stent?"]

cols = [k for k in df.columns if np.any([z in k for z in search_terms])]
df = df[cols]

def rename_types(row):
    v = row["Etiology of SVC Occlusion"]
    if v in name_map.keys():
        return name_map[v]
    else:
        return v
    
def rename_types_az(row):
    v = row["Azizi Type "]
    if v in name_map.keys():
        return name_map[v]
    else:
        return v
    
df["Etiology of SVC Occlusion"] = pu.new_column_from_row_function(df,rename_types)
df["Azizi Type "] = pu.new_column_from_row_function(df,rename_types_az)


import pandas_utils as pu
df = pu.rename_columns(df,{'Emergent Indication (HA, dizziness, syncope, cerebral edema, laryngeal edema, hemodynamic instability, renal insufficiency)':
                           "Emergent Indication"})

df.loc[df.query("Ethnicity == 'White'").index,"Ethnicity"] = "White/Caucasian"


from copy import copy
df_filt = pu.filter_away_nan_rows(df)

from sklearn import preprocessing


encoding_col = ["Emergent Indication","Stenosis Group","Azizi Classification"]
tranformers = dict()
for k in encoding_col:
    le = preprocessing.LabelEncoder()
    le.fit(df_filt[k])
    df_filt[k] = le.transform(df_filt[k])
    tranformers[k] = copy(le)
    
one_hot_columns = ["Ethnicity","Etiology of SVC Occlusion","Azizi Type "]
new_dfs = [pd.get_dummies(df_filt[[k]],drop_first=False) for k in one_hot_columns]

    
def rename_func(row):
    return int(row["Stanford Classification"].split(" ")[-1])

def rename_func_grade(row):
    x = dict(C=0,B=1,A=2)
    return x[row["Azizi Grade"]]

df_filt["Stanford"] = pu.new_column_from_row_function(df_filt,rename_func)
df_filt["Azizi Grade"] = pu.new_column_from_row_function(df_filt,rename_func_grade)
df_filt = pd.concat([df_filt] + new_dfs,axis = 1)
df_filt = pu.delete_columns(df_filt,["Stanford Classification","Azizi Classification"] + one_hot_columns)

pu.set_pd_display_no_truncate()
df_filt
```

# Running the Classifier

```python
import sklearn_models as sklm
clf = sklm.DecisionTreeClassifier(max_depth=2)

import pandas_ml as pdml
X,y = pdml.X_y(df_filt,target_name="Stent?")
clf.fit(X,y)

clf.score(X,y)
```

```python
vml.plot_decision_tree(clf,feature_names=list(X.columns))
```
